# Feature Engineering

```python
# More information about the dataset
df_final.info()
```

 Here many of the features in dataset which are have to be in "Integer" are in "object" or "String" Datatypes. So
 we have to change them all to appropriate datatypes. Now we will be work on every feature of dataset one by one.

## 1. Date_of_Journey

```python
# Date_of_Journey feature
df_final['Date_of_Journey']
```

```python
# More information about the "Date_of_Journey"
df_final['Date_of_Journey'].info()
```

Here we can see that all of the data in this column is in "Object" Datatype. So we convert it into Integer and going to make three new fields from it.
They are: (1)Day    (2)Month    (3)Year

```python
# use split()
df_final['Date'] = df_final['Date_of_Journey'].str.split('/').str[0]
df_final['Month'] = df_final['Date_of_Journey'].str.split('/').str[1]
df_final['Year'] = df_final['Date_of_Journey'].str.split('/').str[2]
```

```python
# Check the results
# Top 5 records of dataset
df_final.head()
```

Here we can see that at last of the dataset, now we have three new columns --> Date, Month, Year  But they are still in "object" and we have to convet it into "Int".

```python
# Convert Object into Integer
df_final['Date'] = df_final['Date'].astype(int)
df_final['Month'] = df_final['Month'].astype(int)
df_final['Year'] = df_final['Year'].astype(int)
```

```python
# More information about the dataset
df_final.info()
```

Now Date, Month and Year are in "Integer" Datatype. We can also drop "Date_of_Journey" column.

```python
# Droping 'Date_of_journey' column from dataset
df_final.drop('Date_of_Journey',axis = 1, inplace = True)
```

```python
# Check the result
# Top 5 records of dataset
df_final.head()
```

"Date_of_Journey" is gone.

## 2. Arrival_Time

```python
# Top 5 records of dataset
df_final.head()
```

In "Arrival_Time" column, we have time and date with month in some cases. But we don't need date with it. The Arrival_Time should be contain only time. Also here the datatype of "Arrival_Time" is Object and we have to convert it into Integer.

```python
# Use split()
df_final['Arrival_Time'] = df_final['Arrival_Time'].str.split(' ').str[0]
```

```python
# More information about the "Arrival_Time"
df_final['Arrival_Time']
```

Now Date from each Arrival_Time is gone. Also, let's make two new features in dataset representing Arrival_Time into hours nad minutes.

```python
# Use split()
df_final['Arrival_Hour'] = df_final['Arrival_Time'].str.split(':').str[0]
df_final['Arrival_Min'] = df_final['Arrival_Time'].str.split(':').str[1]
```

```python
# Check the results
# Top 5 records of dataset
df_final.head()
```

Here we can two new columns "Arrival_Hour" and "Arrival_Min" at last of the dataset. But they are still in Object datatype and we have to convert it into Integer.

```python
# Convert Object into Integer
df_final['Arrival_Hour'] = df_final['Arrival_Hour'].astype(int)
df_final['Arrival_Min'] = df_final['Arrival_Min'].astype(int)
```

```python
# More information about the dataset
df_final.info()
```

Now both the "Arrival_Hour" and "Arrival_Min" columns are in Integer datatype. Now we can also drop "Arrival_Time" column from the dataset as we dont need it.

```python
# Droping 'Arrival_Time' column from dataset
df_final.drop('Arrival_Time',axis = 1, inplace = True)
```

```python
# Check the result
# Top 5 records of dataset
df_final.head()
```

Now "Arrival_time" column is gone from dataset.

## 4. Route and Total_Stops 

```python
# Top 5 records of the dataset
df_final.head()
```

Here in the dataset we have "Source", "Destination" and "Route". In Route "-->" is used to specify the different stops in journey if it is available. We have Source and Destination of jounery and Total Stops in journey already it dataset so, we gonna work on "Route" column and we can drop the "Route" column from dataset. It means No of Routes in journey is equal to No of Total_Stops in journey+1 

```python
# More information about the "Route"
df_final['Route']
```

In this column, the stating point and destination point of journey is given with some stops if it available in between the route of journey.

```python
# More information about the "Total_Stops"
df_final['Total_Stops']
```

But before droping the "Route" column we have to work on "Total_Stops" column.

```python
# Unique values in "Total_Stops" column
df_final['Total_Stops'].unique()
```

```python
# Let's get value for that one 'nan' value in Total_Stops column.
df_final[df_final['Total_Stops'].isnull()]
```

As common sense we can map nan with 2, that is we can assume that there should be atleast 2 stops between Delhi to Cochin.

```python
#Use map() to assign new values 
df_final['Total_Stops'] = df_final['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'nan':2,'4 stops':4})
```

```python
# Check the result
# Top 5 records of the dataset
df_final.head()
```

Here "Total_Stops" get update with new values. NOw we can drop "Route" column from the dataset.

```python
# Use drop()
# Droping 'Route' column from dataset
df_final.drop('Route',axis = 1, inplace = True)
```

```python
# Top 5 records of dataset
df_final.head()
```

Here the "Route" column is gone from the dataset.
